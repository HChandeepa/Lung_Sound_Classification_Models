```python
# Function to perform pitch shifting
def pitch_shift(audio_file, output_file, sr=22050, n_steps=1):
    try:
        # Load the audio file
        y, sr = librosa.load(audio_file, sr=sr)

        # Apply pitch shifting
        y_shifted = librosa.effects.pitch_shift(y, sr=sr, n_steps=n_steps)

        # Write the augmented file
        sf.write(output_file, y_shifted, sr)
        print(f"Augmented file saved: {output_file}")
    except Exception as e:
        print(f"Error processing {audio_file}: {e}")
```

```python
# Function to augment audio samples in a directory
def augment_samples(directory, target_count):
    files = [f for f in os.listdir(directory) if f.endswith('.wav')]  # Filter only '.wav' files
    current_count = len(files)

    # Augment files until we reach the target count
    while current_count < target_count:
        file_to_augment = random.choice(files)

        # Get base name and extension of the chosen file
        base_name, ext = os.path.splitext(file_to_augment)

        # Skip files that have already been augmented (those with '_aug' in the name)
        if "_aug" in base_name:
            continue

        # Create the new augmented file name
        augmented_file = os.path.join(directory, f"{base_name}_aug{current_count}{ext}")

        # Perform pitch shifting augmentation
        pitch_shift(os.path.join(directory, file_to_augment), augmented_file)

        # Update the count after saving the new file
        current_count += 1

    print(f"Augmentation complete. Total files in {directory}: {current_count}")
```
